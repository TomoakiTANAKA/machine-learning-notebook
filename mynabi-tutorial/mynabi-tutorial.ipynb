```python
# imortをつかってライブラリを読み込む
import pandas as pd

allData = pd.read_csv('data/consumerPrices.csv')

allData
```

```python
data1 = allData[['都道府県', '食料', '水道光熱費']]

data1.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.t(data1['食料'], data1['水道光熱費'])
plt.xlabel('Food')
plt.ylabel('Fuel,Light,Water charges')
```

```python
from sklearn.cluster import KMeans

model1 = KMeans(n_clusters=10, random_state=0)
data1_X = data1[['食料', '水道光熱費']]
model1.fit(data1_X)
y1 = model1.labels_

print(y1)
len(y1)
```

```python
data1_results = data1.copy()

data1_results['分類結果'] = y1

data1_results.sort_values('分類結果')
```

```python
plt.scatter(data1_results['食料'], data1_results['水道光熱費'], c=data1_results['分類結果'])
# plt.scatter(data1_results['食料'], data1_results['水道光熱費'])


plt.xlabel('Food')
plt.ylabel('Fuel,Light,Water charges')
```

```python
data2 = allData.copy()
data2_X = data2.drop('都道府県', axis=1)
data2_X.head()
```

```python
model2 = KMeans(n_clusters=4, random_state=0)
model2.fit(data2_X)
y2 = model2.labels_
data2_results = data2.copy()
data2_results['分類結果'] = y2
data2_results.sort_values('分類結果')
```

```python
plt.scatter(data2_results['食料'], data2_results['水道光熱費'], c=data2_results['分類結果'])
plt.xlabel('Food')
plt.ylabel('Fuel,Light,Water charges')
```
